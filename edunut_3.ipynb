```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('SupplyChainEmissionFactorsforUSIndustriesCommodities.csv')
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_").str.replace("[^A-Za-z0-9_]", "", regex=True)
df
```

```python
df.drop(columns=["Commodity_Code","Commodity_Name","Unnamed_7"],inplace=True)
```

```python
df['Supply_Chain_Emission_Factors_without_Margins'][0]=0
df
```

```python
substance_map={'carbon dioxide':0, 'methane':1, 'nitrous oxide':2, 'other GHGs':3}
df['Substance']=df['Substance'].map(substance_map)
```

```python
print(df['Unit'].unique())
```

```python
unit_map={'kg/2018 USD, purchaser price':0, 'kg CO2e/2018 USD, purchaser price':1}
df['Unit']=df['Unit'].map(unit_map)
```

```python
df.drop(columns=["DQ_TemporalCorrelation_of_Factors_without_Margins","DQ_GeographicalCorrelation_of_Factors_without_Margins","DQ_DataCollection_of_Factors_without_Margins"],inplace=True)
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
x=df[['Substance','Unit','Supply_Chain_Emission_Factors_without_Margins','Margins_of_Supply_Chain_Emission_Factors','DQ_ReliabilityScore_of_Factors_without_Margins','DQ_ReliabilityScore_of_Factors_without_Margins','DQ_TechnologicalCorrelation_of_Factors_without_Margins']]
y=df[['Supply_Chain_Emission_Factors_with_Margins']]
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import time
```

```python
n_estimators=100
xgb = XGBRegressor(n_estimators=n_estimators, random_state=42)
```

```python
# Measure training time for XGBoost
start_time_xgb = time.time()
xgb.fit(X_train, y_train)
end_time_xgb = time.time()
xgb_train_time = end_time_xgb - start_time_xgb
```

```python
# Measure prediciton time for XGBoost
start_time_xgb = time.time()
y_pred_xgb = xgb.predict(X_test)
end_time_xgb = time.time()
xgb_pred_time = end_time_xgb - start_time_xgb

mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)


print(f'      XGBoost:  MSE = {mse_xgb:.4f}, R^2 = {r2_xgb:.4f}')
print(f'      XGBoost:  Training Time = {xgb_train_time:.3f} seconds, Testing time = {xgb_pred_time:.3f} seconds')
std_y = np.std(y_test)
```

```python
# XGBoost plot
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_xgb, alpha=0.5, color="orange",ec='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2,label="perfect model")
plt.plot([y_test.min(), y_test.max()], [y_test.min() + std_y, y_test.max() + std_y], 'r--', lw=1, label="+/-1 Std Dev")
plt.plot([y_test.min(), y_test.max()], [y_test.min() - std_y, y_test.max() - std_y], 'r--', lw=1, )
plt.ylim(0,6)
plt.title("XGBoost Predictions vs Actual")
plt.xlabel("Actual Values")
plt.legend()
plt.tight_layout()
plt.show()
```
